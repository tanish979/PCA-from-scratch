```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


n_samples = 100


independent_features = np.random.randn(n_samples, 6)  
columns_independent = [f'X{i+1}' for i in range(6)]


coefficients = np.random.rand(6, 4)  
dependent_features = np.dot(independent_features, coefficients) + np.random.randn(n_samples, 4) * 0.1
columns_dependent = [f'X{i+7}' for i in range(4)]


data = pd.DataFrame(np.hstack([independent_features, dependent_features]),
                    columns=columns_independent + columns_dependent)

```

```python
data
```

```python
X_Transpose = np.array(data)
X_bar_Transpose = X_Transpose - np.mean(X_Transpose,axis = 0)
X_bar_Transpose
```

```python
N_Train =data.shape[0]
X_bar = X_bar_Transpose.T
Sigma = (1/N_Train)*np.matmul(X_bar,X_bar_Transpose)
```

```python
Sigma
```

```python
factorized_sigma = np.linalg.svd(Sigma)
factorized_sigma
```

```python
Q_Transpose = factorized_sigma[0]
Q = Q_Transpose.T
z = np.matmul(Q,X_bar)
z
```

```python
column_names = [f'Z{i+1}' for i in range(z.shape[0])]
projected_training_data = pd.DataFrame(data=z.T, columns=column_names)
projected_training_data
```

```python

Eigen_Values = factorized_sigma[1]
TOT = np.sum(Eigen_Values)
var_sum = 0 
Predefined_variance_percentage = 0.9
for i in range(Eigen_Values.shape[0]):
    var_sum += Eigen_Values[i]
    if (var_sum / TOT) > Predefined_variance_percentage:
        break
projected_dim_reduce_data = projected_training_data.iloc[:, :i]

projected_dim_reduce_data
```
